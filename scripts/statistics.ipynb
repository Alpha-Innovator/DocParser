```python
import os
from collections import defaultdict
from vrdu import utils

dataset_path = "/cpfs01/shared/ADLab/datasets/arxiv_source/arxiv_source_uncompressed/"

result = defaultdict(int)
processed_path = []


for root, dirs, files in os.walk(dataset_path):
    if "layout_annotation.json" not in files:
        continue

    result["processed"] += 1
    processed_path.append(root)

    layout_annotation = utils.load_json(
        os.path.join(root, "layout_annotation.json")
    )
    cur = {
        i: len(
            [x for x in layout_annotation["annotations"] if x["category_id"] == i]
        )
        for i in range(15)
    }
    for key, value in cur.items():
        result[key] += value
```

# generate csv for equation dataset

```python
import os
import pandas as pd
from vrdu import utils

total_df = pd.read_csv("./category_count.csv")

main_path = "/cpfs01/shared/ADLab/datasets/vrdu_equation_final_2"

database_df = pd.DataFrame(
    columns=[
        "quality",
        "date",
        "discpline",
        "image_path",
        "image_height",
        "image_width",
        "paper_source",
        "source_code",
    ]
)

discplines = list(total_df["categories"].values)

for discpline in discplines:
    print(f"processing discpline: {discpline}")
    discpline_path = os.path.join(main_path, discpline)
    json_file = os.path.join(discpline_path, f"reading_annotation_{discpline}.json")
    if not os.path.exists(json_file):
        raise FileNotFoundError(f"{json_file} doesn't exists")

    json_data = utils.load_json(json_file)

    results = [
        {
            "quality": equation_data.get("quality", None),
            "date": equation_data.get("added_date", None),
            "discpline": equation_data.get("discpline", None),
            "image_path": equation_data.get("image_path", None),
            "image_height": equation_data.get("height", None),
            "image_width": equation_data.get("width", None),
            "paper_source": equation_data.get("paper_source", None),
            "source_code": equation_data.get("source_code", None),
        }
        for equation_data in json_data
        if equation_data["image_path"]
    ]

    current_df = pd.DataFrame(results)
    current_df.to_csv(os.path.join(discpline_path, f"equation_database_{discpline}.csv"))
    database_df = pd.concat([database_df, current_df], ignore_index=True)

database_df.to_csv(os.path.join(main_path, f"equation_database.csv"))
```

## Stage 2 result analysis

```python
import os
dataset_path = "/cpfs01/shared/ADLab/ADLab_hdd/vrdu_arxiv/vrdu_autolabel_final_2"

error_type = {
    "missing_layout": [],
    "missing_reading": [],
    "missing_order": [],
    "missing_page": [],
    "missing_original_page": [],
}

for directory in os.listdir(dataset_path):
    paper_path = os.path.join(dataset_path, directory)
    if not os.path.isdir(paper_path):
        continue
    files = os.listdir(paper_path)
    if "layout_annotation.json" not in files:
        error_type["missing_layout"].append(paper_path)
    if "reading_annotation.json" not in files:
        error_type["missing_reading"].append(paper_path)
    if "order_annotation.json" not in files:
        error_type["missing_order"].append(paper_path)
    if "page_0000.jpg" not in files:
        error_type["missing_page"].append(paper_path)
    if "original-page-0001.jpg" not in files:
        error_type["missing_original_page"].append(paper_path)
```

```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_file = "../data.csv"
df = pd.read_csv(data_file)
df.head()
```

```python
discplines = set(df["discpline"])
len(discplines)
```

```python
summary = pd.DataFrame(columns=["discpline", "total", "success", "success_rate"])
for discpline in discplines:
    summary.loc[len(summary)] = [
        discpline,
        len(df[df["discpline"] == discpline]),
        len(df[(df["discpline"] == discpline) & (df["status"] != "failure")]),
        0,
    ]
summary["success_rate"] = summary["success"] / summary["total"]
summary
```

```python
# add a summary row
overall_result = [
    "TOTAL",
    summary["total"].sum(),
    summary["success"].sum(),
    summary["success"].sum() / summary["total"].sum(),
]
print(overall_result)
```

```python
# visualization
# Create an array of indices for positioning the bars
import numpy as np


# Get the unique discplines
discplines = summary["discpline"]

# Create an array of indices for positioning the bars
bar_indices = range(len(discplines))

# Plotting a horizontal bar chart for each category
plt.barh(bar_indices, summary["total"], label="Total")
plt.barh(bar_indices, summary["success"], label="Processed")

# Set y-axis ticks and labels
plt.yticks(bar_indices, discplines, fontsize=6)

# Add labels and title
plt.xlabel("Number of Files")
plt.ylabel("Category")
plt.title("Summary (2w-30w)")

# Add legend
plt.legend()

# Add percentage labels to the bars
for i, total, success in zip(bar_indices, summary["total"], summary["success"]):
    percentage = success / total * 100
    plt.text(total + 2, i, f"{percentage:.1f}%", ha="left", va="center")


# Show the bar chart
plt.savefig("result.png")
plt.show()
```

```python
value_error_info = {
    "title_error": 0,
    "pdflatex_error": 0,
    "latexpand_error": 0,
    "header_error": 0,
    "maxargs_error": 0,
}

for index, row in df[df["error_type"] == "ValueError"].iterrows():
    if "more than one title found" == row["error_info"]:
        value_error_info["title_error"] += 1
    elif "pdflatex" in row["error_info"]:
        value_error_info["pdflatex_error"] += 1
    elif "latexpand" in row["error_info"]:
        value_error_info["latexpand_error"] += 1
    elif "Document class" in row["error_info"]:
        value_error_info["header_error"] += 1
    elif "maxargs_error":
        value_error_info["maxargs_error"] += 1

value_error_info
```

```python
sum(value_error_info.values())
```

```python
filenotfound_error_info = {
    "compile_error": 0,
    "image_pair_error": 0,
}

for index, row in df[df["error_type"] == "FileNotFoundError"].iterrows():
    if "No such file or directory" in row["error_info"]:
        filenotfound_error_info["compile_error"] += 1
    elif "Wrong image path" in row["error_info"]:
        filenotfound_error_info["image_pair_error"] += 1

filenotfound_error_info
```

```python
import pandas as pd

trash_paper_df = pd.read_csv("../trash_paper.csv")
trash_paper_df.head()
```

```python
import os
import glob
import shutil

for index, row in trash_paper_df.iterrows():
    dataset_path = row["path"]
    print(dataset_path)
    paper_id = os.path.basename(dataset_path)[:11]
    result_path = os.path.join(dataset_path, "output/result")

    annotation_images = glob.glob(os.path.join(result_path, "page_*.jpg"))
    for image in annotation_images:
        shutil.copy(
            image,
            os.path.join(
                "/cpfs01/user/maosong/vrdu_data_process/data/",
                f"{paper_id}_{os.path.basename(image)}",
            ),
        )
```

```python
# mark type a error (abstract render)
from vrdu import utils

for index, row in trash_paper_df.iterrows():
    dataset_path = row["path"]
    result_path = os.path.join(dataset_path, "output/result")
    if not os.path.exists(result_path):
        continue
    coco_file = os.path.join(result_path, "layout_annotation.json")
    if not os.path.exists(coco_file):
        continue
    coco_data = utils.load_json(coco_file)

    for annotation in coco_data["annotations"]:
        if annotation["category_id"] == 14:
            prediction_width = annotation["bbox"][2] - annotation["bbox"][0]
            page_width = coco_data["images"][0]["width"]
            if prediction_width > page_width:
                trash_paper_df.loc[index, "error"] = "a"
                break
        elif annotation["category_id"] == 3:
            prediction_height = annotation["bbox"][3] - annotation["bbox"][1]
            page_height = coco_data["images"][0]["height"]
            if annotation["bbox"][1] < 0:
                trash_paper_df.loc[index, "error"] = "b"
                break
```
